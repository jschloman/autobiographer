# Last.fm Listening History Analysis
Deep-dive into your personal musical soundtrack.

```
%load_ext autoreload
%autoreload 2
import os
import sys

import plotly.express as px

# Add the parent directory to sys.path to import analysis_utils
sys.path.append("..")
from analysis_utils import (
    get_hourly_distribution,
    get_listening_intensity,
    get_top_entities,
    load_listening_data,
)
```

## Load Data
Select your CSV file from the `data/` directory.

```
data_file = "../data/lastfm_yourusername_tracks.csv"  # Update this path
if not os.path.exists(data_file):
    print(f"File {data_file} not found. Available files:")
    print([f for f in os.listdir("../data") if f.endswith(".csv")])
else:
    df = load_listening_data(data_file)
    print(f"Loaded {len(df)} tracks.")
    display(df.head())
```

## Statistical Summary

```
if "df" in locals():
    print(f"Total Tracks: {len(df)}")
    print(f"Unique Artists: {df['artist'].nunique()}")
    print(f"Unique Albums: {df['album'].nunique()}")
    if not df.empty:
        print(f"Date Range: {df['date_text'].min()} to {df['date_text'].max()}")
```

## Top Artists

```
if "df" in locals():
    top_artists = get_top_entities(df, "artist", limit=20)
    fig = px.bar(top_artists, x="Plays", y="artist", orientation="h", title="Top 20 Artists")
    fig.update_layout(yaxis={"categoryorder": "total ascending"})
    fig.show()
```

## Hourly Listening Patterns
When do you listen to music the most?

```
if "df" in locals():
    hourly = get_hourly_distribution(df)
    fig = px.bar(hourly, x="hour", y="Plays", title="Listening Intensity by Hour of Day")
    fig.show()
```

## Weekly Trends
Activity over time, aggregated by day.

```
if "df" in locals():
    intensity = get_listening_intensity(df)
    fig = px.line(intensity, x="date", y="Plays", title="Daily Listening Activity")
    fig.show()
```
